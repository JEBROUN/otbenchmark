```python
import openturns as ot
import otbenchmark as otb
import pandas as pd
import numpy as np
```

```python
p = otb.OTReliabilityAlgorithmBenchmark()
List_RP = p.problemslist
np.size(List_RP)
```

Le budget de simulation est Nmax :

```python
Nmax = 10 ** 9
```

```python
N = np.size(List_RP)
model_names = [List_RP[i].getName() for i in range(N)]
metrics = ["Probabilité estimée", "Nombre d'appels à la fonction d'état limite"]
results = np.zeros((len(model_names), len(metrics)))

for i in range(len(model_names)):
    r = otb.OTReliabilityAlgorithmBenchmark.MonteCarloSampling(List_RP[i], maximumOuterSampling=Nmax,
                                                                   coefficientOfVariation=0.1, blockSize=2)
    results[i][0] = np.format_float_scientific(r[0])
    results[i][1] = r[3]
        

df = pd.DataFrame(results, index = model_names, columns=metrics)
df
```

```python
df.to_csv('referenceProba1.csv')
```
